```python
%reload_ext autoreload

import logging
import detaxa.taxonomy as t
import pandas as pd
import numpy as np
import os

logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s [%(levelname)s] %(module)s: %(message)s',
    datefmt='%Y-%m-%d %H:%M',
)

t.loadTaxonomy()
```

```python
import pandas as pd

df = pd.read_csv('data/bacterial_fungal_interactions.taxid.csv')
df
```

```python
df_temp = pd.read_csv('data/bacterial_taxonomy.csv')
df = df.merge(df_temp[['Bacterial_associate_taxa', 'Phylum']].drop_duplicates(), on='Bacterial_associate_taxa', how='left').rename(columns={'Phylum': 'Bacterial_phylum'})
df_temp = pd.read_csv('data/fungal_taxonomy.csv')
df = df.merge(df_temp[['Fungal_host_taxa', 'Phylum']].drop_duplicates(), on='Fungal_host_taxa', how='left').rename(columns={'Phylum': 'Fungal_phylum'})

df = df.fillna('na')

df
```

```python
def name2taxid(name, expand=False, fuzzy=False, reset=False, sk=None):
    x = t.name2taxid(name, expand=expand, max_matches=3, fuzzy=fuzzy, reset=reset, superkingdom=sk)
    if len(x):
        return x[0]
    else:
        return np.nan
```

```python
df['Bacterial_associate_taxa_taxid'] = df.Bacterial_associate_taxa.apply(name2taxid)
df['Fungal_host_taxa_taxid'] = df.Fungal_host_taxa.apply(name2taxid)
```

```python
idx = df.Bacterial_associate_taxa_taxid.isna()
df.loc[idx, 'Bacterial_associate_taxa_taxid'] = df.loc[idx, 'Bacterial_associate_taxa'].apply(lambda x: name2taxid(x, expand=True, fuzzy=True, reset=True))

idx = df.Fungal_host_taxa_taxid.isna()
df.loc[idx, 'Fungal_host_taxa_taxid'] = df.loc[idx, 'Fungal_host_taxa'].apply(lambda x: name2taxid(x, expand=True, fuzzy=True, reset=True, sk='Eukaryota'))
```

```python
df_taxa_idx = pd.DataFrame(df.Bacterial_associate_taxa.unique().tolist() + 
                           df.Fungal_host_taxa.unique().tolist() +
                           df.Bacterial_phylum.unique().tolist() +
                           df.Fungal_phylum.unique().tolist()
                          ).drop_duplicates().reset_index().rename(columns={0: 'name', 'index': 'id'})
df_taxa_idx = df_taxa_idx.fillna('na')
df_taxa_idx
```

```python
df = df.merge(df_taxa_idx, how='left', left_on='Bacterial_associate_taxa', right_on='name').rename(columns={'id': 'bact_id'}).drop(columns=['name'])
df = df.merge(df_taxa_idx, how='left', left_on='Fungal_host_taxa', right_on='name').rename(columns={'id': 'fung_id'}).drop(columns=['name'])
df = df.merge(df_taxa_idx, how='left', left_on='Bacterial_phylum', right_on='name').rename(columns={'id': 'bact_phylum_id'}).drop(columns=['name'])
df = df.merge(df_taxa_idx, how='left', left_on='Fungal_phylum', right_on='name').rename(columns={'id': 'fung_phylum_id'}).drop(columns=['name'])
df
```

```python
df.to_csv('data/bacterial_fungal_interactions.taxid.csv', index=False)
```

```python
df_nodes = df.groupby('Bacterial_associate_taxa').agg({
                        'Fungal_host_taxa_taxid': 'count', 
                        'bact_id':'first', 
                        'Bacterial_phylum': 'first',
                    }).reset_index().rename(columns={
                        'Bacterial_associate_taxa': 'name',
                        'Fungal_host_taxa_taxid': 'count', 
                        'bact_id': 'index', 
                        'Bacterial_phylum': 'phylum'
                    })

df_nodes['group'] = 'Bacteria'


df_temp = df.groupby('Fungal_host_taxa').agg({
                        'Bacterial_associate_taxa': 'count', 
                        'fung_id':'first', 
                        'Fungal_phylum': 'first'
                    }).reset_index().rename(columns={
                        'Fungal_host_taxa': 'name',
                        'Bacterial_associate_taxa': 'count', 
                        'fung_id': 'index', 
                        'Fungal_phylum': 'phylum'
                    })

df_temp['group'] = 'Fungi'

df_nodes = pd.concat([df_nodes, df_temp])

df_nodes = df_nodes.drop_duplicates()

df_nodes['index'] = df_nodes['index'].astype(int)

df_nodes
```

```python
df_links = df[['bact_id', 'fung_id', 'Found_in_16S-CC', 'Found_in_previous_work']].drop_duplicates().rename(columns={
            'bact_id': 'source', 
            'fung_id': 'target', 
            'Found_in_16S-CC': 'found_in_16s',
            'Found_in_previous_work': 'found_in_prev'
            })
df_links['source'] = df_links['source'].astype(int)
df_links['target'] = df_links['target'].astype(int)

df_links
```

```python
df_links = df_links.merge(df_nodes[['index','name']], left_on='source', right_on='index', how='left').rename(columns={'name': 'source_name'}).drop(columns=['index'])
df_links = df_links.merge(df_nodes[['index','name']], left_on='target', right_on='index', how='left').rename(columns={'name': 'target_name'}).drop(columns=['index'])
df_links
```

```python
data = {}

data['nodes'] = df_nodes.sort_values('index').to_dict(orient='records')
data['links'] = df_links.to_dict(orient='records')

import json
# the json file where the output must be stored
out_file = open("data/bacterial_fungal_interactions.vega.json", "w")
   
json.dump(data, out_file)

out_file.close()
```
